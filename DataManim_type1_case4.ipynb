## 의료 비용 예측 데이터



- 데이터 출처 :https://www.kaggle.com/mirichoi0218/insurance/code(참고, 데이터 수정)
데이터 설명 : 의료비용 예측문제
- Data url :https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/train.csv



```python
DataUrl = 'https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/train.csv'
df = pd.read_csv(DataUrl)
display(df.head(5))
```

Q. 흡연자와 비흡연자 각각 charges의 상위 10% 그룹의 평균의 차이는?

```python
smoker_high = df.loc[df['smoker']=='yes']['charges'].quantile(0.9)
nonsmoker_high = df.loc[df['smoker']=='no']['charges'].quantile(0.9)

print(smoker_high, nonsmoker_high)

smoker_mean = df.loc[(df['smoker']=='yes') & (df['charges']>= smoker_high)]['charges'].mean()
nonsmoker_mean = df.loc[(df['smoker']=='no') & (df['charges']>= nonsmoker_high)]['charges'].mean()


Ans = smoker_mean - nonsmoker_mean
Ans
```

## 킹카운티 주거지 가격예측문제

- 데이터 출처 :https://www.kaggle.com/harlfoxem/housesalesprediction (참고, 데이터 수정)
데이터 설명 : 킹카운티 주거지 가격 예측문제
- Data url :https://raw.githubusercontent.com/Datamanim/datarepo/main/kingcountyprice/train.csv

```python
DataUrl = 'https://raw.githubusercontent.com/Datamanim/datarepo/main/kingcountyprice/train.csv'

df = pd.read_csv(DataUrl)
display(df.head(3))
```

Q. bedrooms의 빈도가 가장 높은 값을 가지는 데이터들의 price의 상위 10%와하위 10% 값의 차이를 구하여라

```python
fre_bed = df['bedrooms'].value_counts().index[0]

high_price = df.loc[df['bedrooms']==fre_bed]['price'].quantile(0.9)
low_price = df.loc[df['bedrooms']==fre_bed]['price'].quantile(0.1)

Ans = high_price - low_price
Ans
```

## Q. 대학원 입학가능성 데이터

- 데이터 출처 :https://www.kaggle.com/mohansacharya/graduate-admissions(참고, 데이터 수정)
데이터 설명 : 대학원 입학 가능성 예측
- Data url :https://raw.githubusercontent.com/Datamanim/datarepo/main/admission/train.csv

```python
DataUrl = 'https://raw.githubusercontent.com/Datamanim/datarepo/main/admission/train.csv'
df = pd.read_csv(DataUrl)
display(df.head(3))
```

Q. Serial No. 컬럼을 제외하고 ‘Chance of Admit’을 종속변수, 나머지 변수를 독립변수라 할때, 랜덤포레스트를 통해 회귀 예측을 할 떄 변수중요도 값을 출력하라 (시드값에 따라 순서는 달라질수 있음)

```python
temp = df.drop(columns=['Serial No.'])
display(temp.head(1))
X = temp.drop(columns='Chance of Admit')
display(X.head(1))
y = temp[['Chance of Admit']]
display(y.head(1))
```

```python
from sklearn.ensemble import RandomForestRegressor

ml = RandomForestRegressor()

ml.fit(X,y)
pd.DataFrame({'importance': ml.feature_importances_}, X.columns).sort_values(by='importance', ascending=False)
```
